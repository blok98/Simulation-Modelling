```python
import matplotlib.pyplot as plt
import numpy as np
```

<h4>Growth of amount of fish without restrictions</h4>

```python
def amount_of_fish():
    steps=50
    t=np.zeros(steps+1)
    fish=np.zeros(steps+1)
    fish_diff=np.zeros(steps+1)
    fish[0]=10
    growth_rate=2 #2 fishes born per fish
    for i in range(steps):
        t[i+1]=t[i]+1
        fish_diff[i+1]=fish[i]*growth_rate
        fish[i+1]=fish[i]+  fish_diff[i+1]
    plt.plot(t,fish)
    plt.plot(t,fish_diff)
    plt.show()
```

```python
amount_of_fish()
```

<h4>Growth of amount of fish with maximum capacity</h4>

```python
def amount_of_fish_with_maximum():
    steps=50
    t=np.zeros(steps+1)
    fish=np.zeros(steps+1)
    fish_deviation=np.zeros(steps+1)
    fish[0]=10
    growth_rate=0.5 #fishes born per fish
    carrying_capacity=10**5
    for i in range(steps):
        t[i+1]=t[i]+1
        #fish deviation = 1 - percentage fish of maximum, times the amount of fish
        fish_deviation[i+1] = growth_rate*(1- (fish[i]/carrying_capacity))*fish[i]
        fish[i+1]=fish[i]+  fish_deviation[i+1]
        
    plt.plot(t,fish)
    plt.plot(t,fish_deviation)
    plt.show()
```

```python
amount_of_fish_with_maximum()
```

Aantal vis zal rustig toenemen, omdat de toename afhangt van de hoeveelheid vis. Hoe meer vis er is, hoe hoger de toename. Maar de toename gaat omlaag als het aantal vis de maximum capacity bereikt. Dat komt door de formule <i> 1-(hoeveelheid_vis / maximum_capacity) * hoeveelheid vis</i>. De growth factor gaat omlaag naarmate de maximum_capacity wordt bereikt maar gaat ook omhoog met de hoeveelheid vis. 

<h4>Growth of amount of fish including harvesting</h4>

```python
def amount_of_fish_with_harvesting(harvesting_rate):
#   steps=20
    end_time=10
    h=0.1
    steps=int(end_time/h)
    t=np.zeros(steps+1)
    fish=np.zeros(steps+1)
    fish_deviation=np.zeros(steps+1)
    fish[0]=2e5
    growth_rate=0.5 #fishes born per fish
    carrying_capacity=2e6
    for i in range(steps):
        if fish[i]<=0:
            fish[i+1]=0
            continue
        t[i+1]=t[i]+h
        #fish deviation = 1 - percentage fish of maximum, times the amount of fish
        fish_deviation[i+1] = h*(growth_rate*(1- (fish[i]/carrying_capacity))*fish[i] - harvesting_rate)
        fish[i+1]=fish[i]+  fish_deviation[i+1]
        #if fish are extinct no harvesting nor reproduction will take place
        
        
    return t,fish
```
