```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def f(x):
    return math.e**x
```

<b>Euler-integratie</b><br>
&nbsp;  met beginwaarde f(0)=1, waarbij f(x) = f'(x)


   

```python
def euler(n):
    h=1/n
    
    x = np.zeros(n+1)
    f = np.zeros(n+1)
    f[0] = 1
    
    for i in range(n):
        deviation = math.e**x[i]
        x[i+1]=x[i]+h
        f[i+1]=f[i] + deviation*h

    print("n="+str(n),"f="+str(round(f[-1],4)))
    plt.plot(f)
    plt.show()
    return f
```

```python
for i in [5,10,20,100]:
    euler(i)
```

```python
print(math.e)
difference = math.e-2.7097
difference
```

With 100 steps we reach 2,7097. The difference from e is 0.086.

<b>Heun-integratie</b><br>
&nbsp;  met beginwaarde f(0)=1, waarbij f(x) = f'(x)


   

Now we assume the deviation will change over a change from x. By paying attention to the deviation of the deviation we can estimate the outcome better. The average of the deviation on step i and the deviation on step i+1 will be added up to the previous outcome. 

```python
def heun(n):
    h=1/n
    
    x = np.zeros(n+1)
    f = np.zeros(n+1)
    f[0] = 1
    
    for i in range(n):
        deviation1 = math.e**x[i]
        x[i+1]=x[i]+h
        f[i+1]=f[i] + deviation1*h
        
        deviation2 = math.e**x[i+1]
        average_deviation = (deviation1+deviation2)/2
        f[i+1]=f[i] + average_deviation*h
        
    print("n="+str(n),"f="+str(round(f[-1],5)))
    plt.plot(f)
    plt.show()
    return f
```

```python
for i in [5,10,20,100]:
    (heun(i))
```

```python
print(math.e)
difference = math.e-2.7183
difference
```

With 100 steps we reach 2,7183. The difference from e is only 0.000018!

<h4> Conclusion</h4><br>
Using the Heun method is very accurate to estimate an outcome of a function. Also it is more accurate than the euler method. This is because the Heun method looks forward and how de deviation changes, while the euler method assumes the deviation only changes at every step. A strong change in deviation, or a small stepsize will lead to different outcomes between the 2 methods.
